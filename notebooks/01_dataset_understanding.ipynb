```python
import pandas as pd
```

```python
students = pd.read_csv("../data/raw/students.csv")

courses = pd.read_csv("../data/raw/courses.csv")

instructors = pd.read_csv("../data/raw/instructors.csv")

enrollments = pd.read_csv("../data/raw/enrollments.csv")

quiz_results = pd.read_csv("../data/raw/quiz_results.csv")
```

```python
datasets = {
    "Students": students,
    "Courses": courses,
    "Instructors": instructors,
    "Enrollments": enrollments,
    "Quiz Results": quiz_results
}
```

```python
for name, df in datasets.items():
    print("=" * 60)
    print(name)
    print("=" * 60)

    print("Shape :", df.shape)
    print()

    print(df.info())
```

```python
for name, df in datasets.items():
    print("=" * 60)
    print(name)
    print("=" * 60)

    print(df.isnull().sum())
    print()
```

```python
for name, df in datasets.items():
    print("=" * 60)
    print(name)
    print("=" * 60)

    print("Duplicate :", df.duplicated().sum())
    print()
```
